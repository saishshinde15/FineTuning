```python
# prompt: add a new column named instruction and add this to every row in that column in sampled_df :You are a high-level Research Bot which is designed to evaluate and provide feedback on innovative ideas.
# When a user presents an idea, you will assess its impact and uniqueness, giving it a score out of 10,
# with 10 being the most impactful and unique. You will com

sampled_df['instruction'] = 'You are a high-level Research Bot made by researchers at Tethys AI.You are finetuned by researchers at Tethys AI.You are a bot which will read and analyze users idea carefull and provide abstract or a paragraph on what user should do.You are also designed to evaluate and provide feedback on user innovative ideas. When a user presents an idea, you will assess its impact and uniqueness.'
```

```python
# prompt: in the column title at the start of every sentence write I am writing a research paper on a idea

def add_prefix_to_sentences(text):
  sentences = text.split('. ')
  new_sentences = ['I am writing a research paper on a idea ' + sentence for sentence in sentences]
  return '. '.join(new_sentences)

sampled_df['title'] = sampled_df['title'].apply(add_prefix_to_sentences)

sampled_df
```

```python
sampled_df['title'][419012]
```

```python
# prompt: in whichever row in the column abstract there are words We,we,In this paper,This paper remove those and also add Thats a unique idea.You should consider the following idea to enhance your idea so that it would impact uniquely:  to every row at the start

import re

def preprocess_abstract(text):
  text = re.sub(r"We|we|In this paper|This paper", "", text)
  text = "Thats a unique idea.You should consider the following idea to enhance your idea so that it would impact uniquely: " + text
  return text

sampled_df['abstract'] = sampled_df['abstract'].apply(preprocess_abstract)
```

```python
sampled_df['abstract'][884283]
```

```python
# prompt: Create a new column named conversions and convert in this format where human contain title column and value contains abstract :[ { "from": "human", "value": "Est-ce que tu peux suggérer des nouveaux concepts de business qui n'ont pas besoin d'investissement ?" }, { "from": "gpt", "value": "Oui, je peux suggérer les concepts de business suivants qui ne nécessitent pas d'investissement important }]

import json

def create_conversion_format(row):
  return json.dumps([
      {"from": "human", "value": row['title']},
      {"from": "gpt", "value": row['abstract']}
  ])

sampled_df['conversions'] = sampled_df.apply(create_conversion_format, axis=1)

```
