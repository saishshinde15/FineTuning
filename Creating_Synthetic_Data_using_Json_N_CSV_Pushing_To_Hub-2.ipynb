```python
import json
from datasets import Dataset
import random

# Load your dataset from the JSON file
json_file = "qna_data.json"

with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# Function to generate prompt variations
def generate_prompt_variants(prompt):
    # List of prompt variations
    prompt_variants = [
        prompt,
        prompt.replace("What", "Can you tell me").replace("?", "?"),
        prompt.replace("Who", "Could you explain who").replace("?", "?"),
        prompt.replace("Tell me", "Explain").replace("?", "?"),
        prompt.replace("How", "In what way").replace("?", "?"),
        prompt.replace("about", "regarding").replace("?", "?"),
        prompt.replace("this model", "the AI model").replace("?", "?"),
        prompt.replace("is", "does").replace("?", "?"),
        prompt.replace("more", "further").replace("?", "?"),
        prompt.replace("related", "connected").replace("?", "?"),
        # Additional variations
        prompt.replace("What is", "Can you explain what").replace("?", "?"),
        prompt.replace("Who is", "Who are").replace("?", "?"),
        prompt.replace("Tell me", "Provide details about").replace("?", "?"),
        prompt.replace("How are", "What is the relationship between").replace("?", "?"),
        prompt.replace("Who", "Which person").replace("?", "?"),
        prompt.replace("this model", "the system").replace("?", "?"),
        prompt.replace("fine-tuned", "optimized").replace("?", "?"),
        prompt.replace("CEO", "Chief Executive Officer").replace("?", "?"),
        prompt.replace("CTO", "Chief Technology Officer").replace("?", "?"),
        prompt.replace("related", "associated").replace("?", "?")
    ]
    return prompt_variants

# Function to transform the dataset into the desired format
def transform_to_custom_format(example):
    transformed_entries = []

    # Generate prompt variations
    prompt_variants = generate_prompt_variants(example["prompt"])
    full_response = example["response"]

    # Split the response into sentences
    sentences = full_response.split(". ")

    # Create the reasoning section (first sentence or first few sentences)
    if len(sentences) > 1:
        reasoning = ". ".join(sentences[:2]) + "."  # Use the first two sentences as reasoning
    else:
        reasoning = sentences[0]  # If only one sentence, use it as reasoning

    # Ensure the reasoning doesn't end with "..." by removing incomplete trailing words
    if reasoning.endswith("..."):
        reasoning = reasoning[:-3].strip() + "."

    # Create the answer section (full response)
    answer = full_response.strip()

    # Format the response with <reasoning> and <answer> tags
    formatted_response = (
        f"<reasoning>\n{reasoning}\n</reasoning>\n"
        f"<answer>\n{answer}\n</answer>"
    )

    # Pair each prompt variant with the formatted response
    for variant in prompt_variants:
        transformed_entries.append({"Prompt": variant, "Response": formatted_response})

    return transformed_entries

# Convert the JSON data into a Hugging Face Dataset
dataset = Dataset.from_list(data)

# Apply the transformation to the dataset
all_transformed_entries = []
for example in dataset:
    all_transformed_entries.extend(transform_to_custom_format(example))

# Shuffle and select a subset of entries (optional: limit to 300 entries)
random.shuffle(all_transformed_entries)
final_data = all_transformed_entries[:300]

# Save the transformed dataset to a new JSON file
output_json_file = "transformed_qna_data1.json"

with open(output_json_file, "w", encoding="utf-8") as f:
    json.dump(final_data, f, indent=4, ensure_ascii=False)

print(f"Transformed dataset saved to {output_json_file}")
```
